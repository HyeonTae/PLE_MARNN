```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.decomposition import PCA
import numpy as np
```

```python
data_name = "copy"
data_length = 60
emb_size = 10
hidden_size = 50
rnn = "lstm"
rnn_num = list(range(1,9,1))
```

```python
for i in rnn_num:
    data = (data_name + "_rand_separator_Ctype4_"+ str(data_length)
        + "_att_with_pos_length_emb" + str(emb_size)
        + "_hidden" + str(hidden_size) + "_" + rnn
        + "_" + str(i) + "/")
    
    encoder_pos_weight = np.load(data + "encoder_pos_weight.npy")
    decoder_pos_weight = np.load(data + "decoder_pos_weight.npy")
    
    print(rnn + "_" + str(i) + " ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
    # --------------------------------------------------------------
    # encoder_pos_weight
    principalComponents = PCA(n_components=2)
    pca1 = principalComponents.fit_transform(encoder_pos_weight)

    fig1 = plt.figure(figsize=(8,8))
    ax1 = fig1.add_subplot(111)
    ax1.set_xlabel('PC1', fontsize = 15)
    ax1.set_ylabel('PC2', fontsize = 15)
    ax1.set_title('encoder_pos_weight', fontsize = 20)
    for j in range(len(pca1)):
        ax1.scatter(pca1[j][0], pca1[j][1], c='k')
        ax1.annotate(str(j),(pca1[j][0], pca1[j][1]), color='k')
    #ax2.legend(bbox_to_anchor=(1, 1))
    ax1.grid()
    plt.show()
    # --------------------------------------------------------------
    # decoder_pos_weight
    pca2 = principalComponents.fit_transform(decoder_pos_weight)

    fig2 = plt.figure(figsize=(8,8))
    ax2 = fig2.add_subplot(111)
    ax2.set_xlabel('PC1', fontsize = 15)
    ax2.set_ylabel('PC2', fontsize = 15)
    ax2.set_title('decoder_pos_weight', fontsize = 20)
    for j in range(len(pca2)):
        ax2.scatter(pca2[j][0], pca2[j][1], c='k')
        ax2.annotate(str(j),(pca2[j][0], pca2[j][1]), color='k')
    #ax2.legend(bbox_to_anchor=(1, 1))
    ax2.grid()
    plt.show()
    print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
```
